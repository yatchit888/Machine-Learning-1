## Random Forest Theory revisited

### Random Forest = Decision Tree + Bagging + Random subsets of features

The Random Forest is a model made up of many `decision trees`. Rather than just simply averaging the prediction of trees (which we could call a **forest**), this model uses two key concepts that gives it the name random:
- Random sampling of training data points when building trees
- Random subsets of features considered when splitting nodes

To be more clear, this takes the idea of a single decision tree, and creates an _ensemble_ model out of hundreds or thousands of trees to reduce the variance. 

Each tree is trained on a random set of the observations, and for each split of a node, only a `subset of the features` are used for making a split. When making predictions, the random forest `averages the predictions` for each of the individual decision trees for each data point in order to arrive at a final classification.

### Bagging

### Random sampling of training observations

- **Training**: each tree in a random forest learns from a **random sample** of the data points. The samples are drawn with replacement, known as **bootstrapping**, which means that some samples will be used multiple times in a single tree. The idea is that by training each tree on different samples, although each tree might have high variance with respect to a particular set of the training data, overall, the entire forest will have lower variance but not at the cost of increasing the bias.

- **Testing**: predictions are made by **averaging the predictions** of each decision tree. This procedure of training each individual learner on different bootstrapped subsets of the data and then averaging the predictions is known as **bagging**, short for **bootstrap aggregating**.

### Random Subsets of features for splitting nodes
Only a subset of all the features are considered for splitting each node in each decision tree. Generally this is set to `sqrt(n_features)` for classification meaning that if there are 16 features, at each node in each tree, only 4 random features will be considered for splitting the node.

### Import RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
```

### Import RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

### Set the parameter grid according to the requirements above as a dictionary

```python
param_grid = {
    'n_estimators': np.arange(10, 200),
    'max_depth': np.arange(3, 20),
    'max_features': ['auto', 'sqrt', None] + list(np.arange(0.5, 1, 0.1)),
    'max_leaf_nodes': np.arange(10, 50),
    'min_samples_split': [2, 5, 10],
    'bootstrap': [True, False]
}
```

### Create the estimator with RSEED

```python
estimator = RandomForestClassifier(random_state = RSEED)
```

### Create the Random Search model with cv=3, n_iter=10, scoring='roc_auc', random_state='RSEED'

```python
model = RandomizedSearchCV(estimator, param_distributions=param_grid, cv=3,  scoring = 'roc_auc', n_iter = 10, random_state = RSEED)
```

### Fit the model 
Note: It will take long time (around 20 - 1 hour depending on your computer specs). Good time to reload yourself with some energy or take a quick beauty nap!!!

```python
model.fit(train, train_labels)
```

### Explore the best parameters

```python
best_params = model.best_params_
print(f"Best Parameters: {best_params}")
```

Choose the best model as you find in under `best_estimator_`

```python
label_pred = best_model.predict(test)
```

### Get the node counts and maximum depth of the random forest

```python
tree_node_counts = [estimator.apply(train).sum(axis=0) for estimator in best_model.estimators_]
tree_max_depths = [estimator.get_depth() for estimator in best_model.estimators_]

for i, (node_count, max_depth) in enumerate(zip(tree_node_counts, tree_max_depths)):
    print(f"Tree {i+1} - Node Counts: {node_count}, Maximum Depth: {max_depth}")
```

## Plot the ROC AUC Scores for training and testing data

```python
from sklearn.metrics import roc_curve, auc
train_probs = best_model.predict_proba(train)[:, 1]
test_probs = best_model.predict_proba(test)[:, 1]
fpr_train, tpr_train, _ = roc_curve(train_labels, train_probs)
roc_auc_train = auc(fpr_train, tpr_train)
fpr_test, tpr_test, _ = roc_curve(test_labels, test_probs)
roc_auc_test = auc(fpr_test, tpr_test)

plt.figure(figsize=(8, 6))
plt.plot(fpr_train, tpr_train, color='darkorange', lw=2, label='Train ROC curve (area = {:.2f})'.format(roc_auc_train))
plt.plot(fpr_test, tpr_test, color='navy', lw=2, label='Test ROC curve (area = {:.2f})'.format(roc_auc_test))
plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

### Helper function to Evaluate model

```python
from sklearn.metrics import roc_auc_score, roc_curve, recall_score, precision_score

def evaluate_model(predictions, probs, train_predictions, train_probs):
    """Compare machine learning model to baseline performance.
    Computes statistics and shows ROC curve."""
    
    baseline = {}
    
    baseline['recall'] = recall_score(test_labels, [1 for _ in range(len(test_labels))])
    baseline['precision'] = precision_score(test_labels, [1 for _ in range(len(test_labels))])
    baseline['roc'] = 0.5
    
    results = {}
    
    results['recall'] = recall_score(test_labels, predictions)
    results['precision'] = precision_score(test_labels, predictions)
    results['roc'] = roc_auc_score(test_labels, probs)
    
    train_results = {}
    train_results['recall'] = recall_score(train_labels, train_predictions)
    train_results['precision'] = precision_score(train_labels, train_predictions)
    train_results['roc'] = roc_auc_score(train_labels, train_probs)
    
    for metric in ['recall', 'precision', 'roc']:
        print(f'{metric.capitalize()} Baseline: {round(baseline[metric], 2)} Test: {round(results[metric], 2)} Train: {round(train_results[metric], 2)}')
    
    # Calculate false positive rates and true positive rates
    base_fpr, base_tpr, _ = roc_curve(test_labels, [1 for _ in range(len(test_labels))])
    model_fpr, model_tpr, _ = roc_curve(test_labels, probs)

    plt.figure(figsize = (8, 6))
    plt.rcParams['font.size'] = 16
    
    # Plot both curves
    plt.plot(base_fpr, base_tpr, 'b', label = 'baseline')
    plt.plot(model_fpr, model_tpr, 'r', label = 'model')
    plt.legend();
    plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate'); plt.title('ROC Curves');
```

### Evaluate the best model

```python
evaluate_model(label_pred, test_probs, best_model.predict(train), train_probs)
```

### Confusion Matrix Helper function

```python
from sklearn.metrics import confusion_matrix
import itertools

#  Helper function to plot Confusion matrix
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Oranges):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    Source: http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.figure(figsize = (10, 10))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title, size = 24)
    plt.colorbar(aspect=4)
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45, size = 14)
    plt.yticks(tick_marks, classes, size = 14)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    
    # Labeling the plot
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt), fontsize = 20,
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
        
    plt.grid(None)
    plt.tight_layout()
    plt.ylabel('True label', size = 18)
    plt.xlabel('Predicted label', size = 18)
```

### What is GridSearchCV, RandomizedSearchCV?

GridSearchCV stands for cross-validation. It evaluates all grid combinations of model hyperparameters to find the optimal one.

RandomizedSearchCV, a GridSearchCV version, doesn't search all hyperparameter combinations. Instead, it selects a set number of hyperparameter combinations from the space. When considering several hyperparameters, this may be more efficient.

### Why there is RCV when GSCV is already there?

GridSearchCV tests every hyperparameter combination. This may be time-consuming, particularly with several hyperparameters and values to consider. A predetermined number of hyperparameter combinations are randomly sampled via RandomizedSearchCV. Sometimes this is more efficient, particularly in huge search spaces.

GridSearchCV searches every possible combination. If the search space contains the ideal combination, it may locate it. In contrast, RandomizedSearchCV randomly examines options. It may not discover the ideal combination, but it can find decent ones faster.

### When to use what CV?

Use GridSearchCV When:

1. Small Hyperparameter Space:
When you have a relatively small number of hyperparameters and a limited set of values to explore, GridSearchCV can be a good choice. It allows you to perform an exhaustive search efficiently.

2. Computationally Feasible
When the number of hyperparameter combinations is manageable given your computational resources, GridSearchCV can be a suitable option.

3. Need for Certainty:
If you have a strong belief that the optimal hyperparameters exist within the specified search space and you want to ensure you've explored every possibility, GridSearchCV is a more deterministic approach.

Use RandomizedSearchCV When:

1. Large Hyperparameter Space:
When the hyperparameter space is very large and performing an exhaustive search is computationally impractical, RandomizedSearchCV is a more efficient alternative.

2. Limited Computational Resources:
If you have limited time or computational resources, RandomizedSearchCV allows you to explore a wide range of hyperparameters without the computational cost of exhaustively evaluating all combinations.

3. Exploratory Analysis:
If you're in the early stages of model development and want to get a sense of which hyperparameters might be important without committing to an exhaustive search, RandomizedSearchCV can provide valuable insights.

4. Finding Good, Not Necessarily Best, Hyperparameters:
If you're more interested in finding very good combinations of hyperparameters rather than necessarily the absolute best, RandomizedSearchCV can be a pragmatic choice.
